```python
# --- SIMULADOR DE SESIÓN PSICOLÓGICA CON IA ---

# Entrada de datos inicial para personalizar la sesión
nombre = input("Bienvenido a la consulta virtual. ¿Cómo te llamas? ")

# Operador relacional para verificar si el nombre fue ingresado
if nombre == "":
    nombre = "Paciente"

print(f"Hola {nombre}, soy tu asistente psicólogo virtual.")
respuesta_inicial = input("¿Cómo te has sentido hoy emocionalmente? (bien/mal/estresado): ").lower()

# Estructura condicional múltiple (if-elif-else) para determinar el diagnóstico
if respuesta_inicial == "bien" or respuesta_inicial == "feliz":
    # Uso de operador lógico 'or' para capturar estados positivos [5]
    print("Es excelente que te sientas así. Mantener una actitud positiva es clave para la salud mental.")

elif respuesta_inicial == "mal" or respuesta_inicial == "triste" or respuesta_inicial == "estresado":
    # Entrada de datos numérica para aplicar operadores aritméticos y relacionales [6, 7]
    nivel_estres = int(input("En una escala del 1 al 10, ¿qué tan intenso es ese sentimiento? "))

    # Evaluación del nivel utilizando operadores relacionales y lógicos [5, 7]
    if nivel_estres >= 8 and nivel_estres <= 10:
        # Cálculo aritmético simple para mostrar un porcentaje de intensidad
        porcentaje = nivel_estres * 10
        print(f"Tu malestar está a un nivel crítico del {porcentaje}%.")
        print("Recomiendo que realices ejercicios de respiración profunda inmediatamente.")

    elif nivel_estres >= 4 and nivel_estres < 8:
        # Rango intermedio de estrés
        print("Es un nivel considerable de estrés. Sería bueno identificar la causa principal de esa presión.")

    elif nivel_estres > 0 and nivel_estres < 4:
        # Nivel bajo de estrés
        print("Parece algo leve. Un poco de descanso o una caminata podrían ayudarte hoy.")

    else:
        # Caso por si el usuario introduce un número fuera del rango 1-10
        print("Has ingresado un valor fuera de la escala, pero entiendo que hay incomodidad.")

else:
    # Bloque de salida por defecto si no se reconoce la emoción
    print("A veces es difícil describir cómo nos sentimos. Lo importante es que estás aquí buscando apoyo.")

# Mensaje final de la sesión
print(f"Gracias por confiar en mí hoy, {nombre}. Recuerda que esta es una simulación básica.")
```
